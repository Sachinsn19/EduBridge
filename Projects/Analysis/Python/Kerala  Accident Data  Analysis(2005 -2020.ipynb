# CONCLUSION

From data analysed based on months, data showing  that most number of accident cases registered ,total persons injured and total deaths were in the month of January. When analyse the data based on year wise,data showing that Most number of accident cases registered, total persons injured and total deaths were in the year of 2005

From data analysed based  on time intervals and type of roads which accident occured, data showing that most number of accident cases registered and most number of accident injuries in the time interval of 3.00PM to 6.00PM,  and most number of accident deaths  occured in the time interval of 6.00PM to 9.00PM. Most number of accident cases ,injuries and deaths were on the roads which are other than National highway and State high ways in Kerala.

The main reason behind every accident cases registered ,injuries and deaths were  the fault of the driver of the vehicle.The percentage of contribution of fault of the driver of vehicle in accident cases ,injuries and deaths were greater than 95%.
